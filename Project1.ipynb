# allow three ascii codes

```python
def is_english(app_name):
    non_asci = 0
    for i in app_name:
        if ord(i) > 127:
            non_asci +=1
    if non_asci > 3:
        return False
    else:
        return True
```

```python
is_english("Instagram")
```

# most common genre as per avg installation

```python
avg_istallation = []
for i in unique_catagories:
    total_ins=0
    no_of_apps=0
    for j in android_final:
        catagory = j[1]
        if i == catagory:
            install = float(j[5].replace("+","").replace(",",""))
            total_ins +=install
            no_of_apps +=1
    avg = total_ins/no_of_apps
#     print(avg)
    avg_istallation.append([i,avg])
```
